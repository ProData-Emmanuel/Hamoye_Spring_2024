```python
#  What is the total Protein supply quantity in Madagascar in 2015?
madagascar_proteinsupply_quantity = df[(df['Area'] == 'Madagascar') & (df['Element'] == 'Protein supply quantity (g/capita/day)')]

# The Sum of the Import Quantity of ALgeria in year 2018:
madagascar_proteinsupply_quantity_2018 = madagascar_proteinsupply_quantity['Y2015'].sum()
madagascar_proteinsupply_quantity_2018
```

```python
# Finding the Areas that had the 7th lowest sum in 2017?

df[['Y2017', 'Area']].groupby('Area').sum()
```

```python
# Grouping by Area and sorting by increasing order of the Y2017
df[['Y2017', 'Area']].groupby('Area').sum().sort_values(by='Y2017')
```

```python
# Which of these elements had the highest sum in 2017?
df[['Y2017', 'Element']].groupby('Element').sum()
```

```python
# Determining the year with the highest sum of Stock Variation?
df.groupby('Element').sum()
```

```python
# Calculating Correlation
relevant_columns = df[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]

# Calculating the Correlations
corre = relevant_columns.corr()
```

```python
# Checking for the Column with the least correlation from the correlation values, shows year 2016 has the least correlation with Element Code
corre
```

```python
# Importing plotting modules
import matplotlib.pyplot as plt
import seaborn as sns

# plotting the correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corre, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
